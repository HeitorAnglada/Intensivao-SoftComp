## Dado

*   Dados são fatos brutos, números, símbolos ou registros que não têm contexto ou significado inerente por si mesmos.
*   Eles são frequentemente coletados de fontes diversas e podem ser representados de várias formas, como texto, números, imagens, sons, etc.
* Os dados são a matéria-prima da informação; eles precisam ser processados e interpretados para se tornarem úteis.    
* Exemplo: Uma lista de números 1, 2, 3, 4, 5 é um conjunto de dados.

##Informação

  * Informações são dados que foram organizados, processados, contextualizados e dotados de significado, tornando-os úteis para tomada de decisões ou entendimento.
  * As informações são o resultado da análise e interpretação dos dados, que permite que as pessoas compreendam um determinado contexto ou situação.
  * Elas têm a capacidade de influenciar ações ou decisões.
  * Exemplo: Se os números 1, 2, 3, 4, 5 representam as notas de um aluno em cinco disciplinas diferentes, as informações podem ser que a média dessas notas é 3.

# O que é Análise Exploratória de Dados?

A análise exploratória de dados, ou AED, é o processo de investigar, resumir e visualizar dados para obter percepções iniciais e compreender melhor as informações contidas neles. É a etapa crucial antes de qualquer análise estatística avançada ou modelagem de dados. Através da AED, podemos responder a perguntas importantes sobre os nossos dados, identificar tendências, detectar outliers e formular hipóteses iniciais. Em resumo, a AED ajuda a tirar o máximo proveito dos dados disponíveis e a tomar decisões informadas.

# Sobre dados


##  Tipos de dados



Dados Categóricos:

* São dados que representam categorias ou grupos distintos.
* Exemplo: Cores (vermelho, verde, azul), tipos de animais (cachorro, gato, pássaro), nomes próprios.

Dados Ordinais:

* São dados que possuem uma ordem intrínseca entre as categorias, mas as diferenças entre elas não são quantificáveis de maneira uniforme.
* Exemplo: Níveis de satisfação (insatisfeito, neutro, satisfeito), grau de educação (ensino fundamental, médio, graduação, pós-graduação).

Dados Contínuos:

* São dados numéricos que podem assumir um número infinito de valores dentro de um intervalo.
* Exemplo: Idade, altura, temperatura, salário.

Esses três tipos de dados são fundamentais na análise de dados e desempenham papéis diferentes na compreensão e interpretação de informações em diferentes contextos.

#Medida de tendencia central e dispersão

##Medidas de tendencia central

Medidas de tendência central são estatísticas usadas para resumir e descrever a localização central de um conjunto de dados. Elas ajudam a entender onde a maioria dos valores se concentra e representam um valor típico ou central do conjunto de dados. As principais medidas de tendência central são:

###Média

 A média é a soma de todos os valores dividida pelo número de observações. É representada pela fórmula: Média = (Soma dos Valores) / (Número de Observações). A média é sensível a valores extremos (outliers).

```python
df.mean()
```

```python
df.describe()
```

### Mediana

A mediana é o valor que divide o conjunto de dados ao meio quando todos os valores estão organizados em ordem crescente. É uma medida de tendência central robusta aos outliers, o que significa que não é afetada por valores extremos.

```python
#median
```

###Moda

A moda é o valor que aparece com mais frequência em um conjunto de dados. Pode haver uma moda (unimodal), mais de uma moda (multimodal), ou nenhum valor que se repita (sem moda).

```python
#mode
```

## Dispersão

### Variância

A dispersão é uma medida geral da variabilidade dos dados, mas não nos dá detalhes específicos sobre como os valores se espalham.

* A variância é uma medida numérica da dispersão que calcula a média dos quadrados das diferenças entre cada valor de dados e a média do conjunto de dados.
* A fórmula da variância populacional é: Var(X) = Σ (Xi - μ)² / N, onde Xi são os valores, μ é a média e N é o número de observações.

```python
#var
```

### Desvio Padrão

O desvio padrão é uma medida que representa a raiz quadrada da variância. Ele fornece uma ideia da dispersão dos dados em unidades originais, o que torna a interpretação mais intuitiva.
* A fórmula do desvio padrão populacional é: Desvio Padrão(X) = √Var(X).

#Quartis

Os quartis são medidas estatísticas que dividem um conjunto de dados ordenados em quatro partes iguais, cada uma representando 25% dos dados. Eles são úteis para entender a distribuição dos dados e identificar valores atípicos. Os quartis dividem os dados em três pontos específicos:

* Primeiro Quartil (Q1): O primeiro quartil é o valor que separa os 25% dos dados inferiores quando os dados estão organizados em ordem crescente. É o valor no qual 25% dos dados estão abaixo e 75% estão acima.

* Segundo Quartil (Q2): O segundo quartil é simplesmente a mediana dos dados, o valor que separa os 50% inferiores dos 50% superiores. É o mesmo que o valor da mediana.

* Terceiro Quartil (Q3): O terceiro quartil é o valor que separa os 75% dos dados inferiores quando os dados estão organizados em ordem crescente. É o valor no qual 75% dos dados estão abaixo e 25% estão acima.

# Outlers

Outliers são valores atípicos em um conjunto de dados que se desviam significativamente da maioria dos outros valores. Eles podem ser dados que estão muito acima ou muito abaixo da média ou mediana do conjunto de dados. Outliers podem ser causados por erros de medição, variações naturais ou eventos excepcionais.

## Boxplot

Um box plot, ou gráfico de caixa, é uma representação gráfica que ajuda a identificar e visualizar outliers em um conjunto de dados. Ele mostra a distribuição dos dados por meio de quartis e fornece uma representação gráfica das medidas estatísticas importantes, como mediana, quartis e possíveis outliers.

```python
#boxplot
df.boxplot()
```
